```python
import pandas as pd
```

#### Load Excel-File and identify Worksheets

```python
path = '../datalake/online_retail_II.xlsx'
xl = pd.ExcelFile(path)
xl.sheet_names  # see all sheet names
```

#### Read Excel-Files to Dataframe

```python
df_1 = pd.read_excel(path, sheet_name='Year 2009-2010')
df_2 = pd.read_excel(path, sheet_name='Year 2010-2011')
df = pd.concat([df_1, df_2], axis=0, join='inner', ignore_index=True)
```

```python
df
```

```python
df_pivot = df.pivot_table(index="Customer ID", values="Country", aggfunc=(lambda x: x))
df_customers = pd.DataFrame({'customer_id': df_pivot.index, 'country': df_pivot['Country']}).reset_index()
df_customers['country'] = df_customers['country'].apply(lambda x: x[0])
df_customers = df_customers[['customer_id', 'country']]
df_customers['customer_id'] = df_customers['customer_id'].astype(int)
df_customers
```

```python
path = "../csv-files/customers_data.csv"
df_customers.to_csv(path, index=False)
```
